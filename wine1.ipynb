```python
with engine.connect() as conn:
    col_exists = conn.execute(text("""
        SELECT COUNT(*) 
        FROM information_schema.columns
        WHERE table_name = 'wine_red'
        AND column_name = 'wine_type'
    """)).scalar()

    if col_exists == 0:
        conn.execute(text("""
            ALTER TABLE wine_red
            ADD COLUMN wine_type VARCHAR(10)
        """))

    conn.execute(text("""
        UPDATE wine_red
        SET wine_type = 'red'
    """))

    conn.commit()
```

```python
with engine.connect() as conn:
    col_exists = conn.execute(text("""
        SELECT COUNT(*) 
        FROM information_schema.columns
        WHERE table_name = 'wine_white'
        AND column_name = 'wine_type'
    """)).scalar()

    if col_exists == 0:
        conn.execute(text("""
            ALTER TABLE wine_white
            ADD COLUMN wine_type VARCHAR(10)
        """))

    conn.execute(text("""
        UPDATE wine_white
        SET wine_type = 'white'
    """))

    conn.commit()
```

```python
with engine.connect() as conn:
    conn.execute(text("DROP TABLE IF EXISTS wine_combined"))

    conn.execute(text("""
        CREATE TABLE wine_combined AS
        SELECT * FROM wine_red
        UNION
        SELECT * FROM wine_white
    """))

    conn.commit()
```

Quality Segmentation Analysis (Low / Medium / High)
* less than 4 treat as low
* between 5 & 6 as medium
* greater than 7 as high

```python
from sqlalchemy import text

with engine.connect() as conn:
    result = conn.execute(text("""
        SELECT
            CASE
                WHEN quality <= 4 THEN 'Low'
                WHEN quality BETWEEN 5 AND 6 THEN 'Medium'
                ELSE 'High'
            END AS quality_level,
            COUNT(*) AS count
        FROM wine_combined
        GROUP BY quality_level
    """))
    print(result.fetchall())
```

Wine Type Ã— Quality Level Ã— Alcohol

```python
with engine.connect() as conn:
    result = conn.execute(text("""
        SELECT wine_type,
               CASE
                   WHEN quality <= 4 THEN 'Low'
                   WHEN quality BETWEEN 5 AND 6 THEN 'Medium'
                   ELSE 'High'
               END AS quality_level,
               AVG(alcohol) AS avg_alcohol
        FROM wine_combined
        GROUP BY wine_type, quality_level
        ORDER BY wine_type, quality_level
    """))
    print(result.fetchall())
```

Volatile Acidity Impact on Quality
* Higher volatile acidity â†’ lower quality
* Your "volatile acidity" values are ALL < 0.3

```python
from sqlalchemy import text

with engine.connect() as conn:
    result = conn.execute(text("""
        SELECT
            CASE
                WHEN "volatile acidity" < 0.3 THEN 'Low'
                WHEN "volatile acidity" BETWEEN 0.3 AND 0.6 THEN 'Medium'
                ELSE 'High'
            END AS acidity_level,
            AVG(quality) AS avg_quality
        FROM wine_combined
        GROUP BY acidity_level
    """))
    print(result.fetchall())
```

Sulphates + Alcohol Combined Effect
* High sulphates + High alcohol = Best quality wines

```python
with engine.connect() as conn:
    result = conn.execute(text("""
        SELECT
            CASE
                WHEN sulphates < 0.5 THEN 'Low Sulphates'
                ELSE 'High Sulphates'
            END AS sulphate_level,
            CASE
                WHEN alcohol < 10 THEN 'Low Alcohol'
                ELSE 'High Alcohol'
            END AS alcohol_level,
            AVG(quality) AS avg_quality
        FROM wine_combined
        GROUP BY sulphate_level, alcohol_level
    """))
    print(result.fetchall())
```

Density Ã— Sugar Ã— Alcohol Relationship
* As alcohol increases:

Density decreases

Sugar impact reduces

```python
from sqlalchemy import text

with engine.connect() as conn:
    result = conn.execute(text("""
        SELECT
            CAST(alcohol AS REAL) AS alcohol_level,
            AVG(CAST("residual sugar" AS REAL)) AS avg_sugar,
            AVG(CAST(density AS REAL)) AS avg_density
        FROM wine_combined
        GROUP BY alcohol_level
        ORDER BY alcohol_level
    """))
    print(result.fetchall())
```
